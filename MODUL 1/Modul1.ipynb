```python
fruits = ["apel", "mangga", "jeruk"]

print(fruits[0]) #mencetak isi array ke [i]
print(fruits[1])

fruits[1] = "pisang" #mengganti isi array dari mangga ke pisang
print(fruits)

for fruit in fruits: #perkondisian loop pencarian isi dahulu
  print(fruit)
if"pisang" in fruits: #perkondisian jika pisang ada dalam array
  print ("Ada pisang!")

print(len(fruits)) #mencetak panjang array
fruits.append("anggur") # menambahkan isian array di belakang indeks terakhir
print(fruits)
```

```python
f = lambda x: x**2
print(f(5))

def fib(n):
    if n <= 1:
        return n
    else:
        return (fib(n-1) + fib(n-2))

print(fib(10))

def gen(n):
    for i in range(n):
        yield i

g = gen(5)
for x in g:
    print(x)
```

```python
import csv
with open("fruits.csv",mode="w",newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name','Age','City'])
    writer.writerow(['Adrian','20','Bandung'])
    writer.writerow(['Ronaldo','37','Sporting Lisbon'])
    writer.writerow(['Pessiun','35','Barcelona'])

with open("fruits.csv", newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
